# Шаг 1 - скачиваем данные с сайта ЦБ

```python
# Создаём директории Directory for archieves, Directory for unarchieved files
path = 'D:/diploma'

try:
    os.mkdir(path + 'Directory for archieves')
except OSError:
    print ("Creation of the directory %s failed" % path + '\Directory for archieves')
else:
    print ("Successfully created the directory %s " % path + '\Directory for archieves')


try:
    os.mkdir(path + "\Directory for unarchieved files")
except OSError:
    print ("Creation of the directory %s failed" % path + "\Directory for unarchieved files")
else:
    print ("Successfully created the directory %s " % path + "\Directory for unarchieved files")
```

```python
# скачиваем все ссылки на различные формы отчетности с сайта ЦБ
url = 'http://cbr.ru/vfs/credit/forms/'
r = requests.get(url)
html_content = r.text
soup = BeautifulSoup(html_content, 'lxml')

raw_links = soup.find_all('a')
links = [i.get('href') for i in raw_links]
links[:10]
```

```python
# выбираем лишь те, что относятся к 101 форме
links = links[3:]
links_101 = [i for i in links if i[18:21] == '101']
links_101[:10]
```

```python
# выбираем лишь те, что относятся к 101 форме и после 2013 г.
link_101_select = []

for link in links_101:
    if int(link[22:26]) > 2013:
        link_101_select.append(link)
```

```python
link_101_select
```

```python
# скачиваем данные
for link in link_101_select:
    urllib.request.urlretrieve(r'http://cbr.ru' + link,
                               path + "/Directory for archieves" + link[17:])
```

# Шаг 2 - распаковываем архивы

```python
# распаковываем данные 
for link in link_101_select:
    os.mkdir(r'D:\diploma\Directory for unarchieved files' + str(link[17:30]))    
    patoolib.extract_archive(path + "/Directory for archieves" + str(link[17:]),
                         outdir = path + "/Directory for unarchieved files" + str(link[17:30]))  
    
                             


```

```python
# переименовываем папки с распакованными данными так, чтобы имя папки совпадало с именем ключевового файла в части месяца и года
for directory in os.listdir(path + "\Directory for unarchieved files"):
    os.rename(path + "\Directory for unarchieved files\\" + directory,  
             path + "\\Directory for unarchieved files\\" + os.listdir(path + "\Directory for unarchieved files\\" + directory)[0][:6])
```

# Шаг 3 - скачиваем данные с сайта ЦБ по валютному курсу

```python
# выбираем текущую дату
now = datetime.datetime.now()

day = str(now.day)
month = str(now.month)
year = str(now.year)

current_date = day + '.' + month + '.' + year + '.'
current_date
```

```python
# скачивам данные с сайта ЦБ по официальному крусу USD/RUB по текущую дату

url = r'https://cbr.ru/currency_base/dynamics/?UniDbQuery.Posted=True&UniDbQuery.so=1&UniDbQuery.mode=1&UniDbQuery.date_req1=&UniDbQuery.date_req2=&UniDbQuery.VAL_NM_RQ=R01235&UniDbQuery.From=01.01.2013&UniDbQuery.To=' + current_date

def get_data(url: str):
    df_er = pd.read_html(url)
    return df_er

df_er = get_data(url)
df_er = df_er[0]
df_er
```

```python
# обрабатываем данные по валютному курсу

df_er = df_er.iloc[2:]     # deleting the first useless row 'Доллар США' and the second row since he will be in the 
                     # column names instead

df_er = df_er.drop(1, axis = 1)   # deleting useless column "Единиц"
df_er = df_er.rename({0: 'date', 2: 'exchange rate'}, axis=1)    # adding normal column names
df_er.iloc[:, 1] = df_er.iloc[:, 1].astype('float')/10000       # 
df_er = df_er.set_index('date')         # setting date as index

df_er
```
